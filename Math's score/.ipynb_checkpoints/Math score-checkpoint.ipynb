Training algorithyms for regression

Use $L_{2}$ loss $L(y, z) = (y-z)^{2},$ it's derevative by $z$ will look like $L'(y, z) = 2(z-y)$. So will implement it as function (can get rid of 2*).

```python
def bias(y, z):
    return (y - z)
```

Mean squared error will be used as a quality indicator.

```python
def mean_squared_error(y_real, prediction):
    return (sum((y_real - prediction)**2)) / len(y_real)
```

```python
def gb_predict(X, trees_list, coef_list, eta):
    #        ,
    #      trees_list          eta
    return np.array([sum([eta* coef * alg.predict([x])[0] for alg, coef in zip(trees_list, coef_list)]) for x in X])
```

```python
def gb_fit(n_trees, max_depth, X_train, X_test, y_train, y_test, coefs, eta):
    
    #     
    trees = []
    
    #             
    train_errors = []
    test_errors = []
    
    for i in range(n_trees):
        tree = DecisionTreeRegressor(max_depth=max_depth, random_state=42)

        #    ,  , 
        #           
        if len(trees) == 0:
            #      
            tree.fit(X_train, y_train)
            
            train_errors.append(mean_squared_error(y_train, gb_predict(X_train, trees, coefs, eta)))
            test_errors.append(mean_squared_error(y_test, gb_predict(X_test, trees, coefs, eta)))
        else:
            #     
            target = gb_predict(X_train, trees, coefs, eta)
            
            #       
            tree.fit(X_train, bias(y_train, target))
            
            train_errors.append(mean_squared_error(y_train, gb_predict(X_train, trees, coefs, eta)))
            test_errors.append(mean_squared_error(y_test, gb_predict(X_test, trees, coefs, eta)))

        trees.append(tree)
        
    return trees, train_errors, test_errors
```

```python
def evaluate_alg(X_train, X_test, y_train, y_test, trees, coefs, eta):
    train_prediction = gb_predict(X_train, trees, coefs, eta)

    print(f'   {n_trees}   {max_depth} \
      {eta}   : {mean_squared_error(y_train, train_prediction)}')

    test_prediction = gb_predict(X_test, trees, coefs, eta)

    print(f'   {n_trees}   {max_depth} \
      {eta}   : {mean_squared_error(y_test, test_prediction)}')
```

```python
#    
n_trees = 15

#      1
coefs = [0.9] * n_trees

#   
max_depth = 3

# 
eta = 0.5

trees, train_errors, test_errors = gb_fit(n_trees, max_depth, X_train, X_test, y_train, y_test, coefs, eta)
```

```python
def get_error_plot(n_trees, train_err, test_err):
    plt.xlabel('Iteration number')
    plt.ylabel('MSE')
    plt.xlim(0, n_trees)
    plt.plot(list(range(n_trees)), train_err, label='train error')
    plt.plot(list(range(n_trees)), test_err, label='test error')
    plt.legend(loc='upper right')
    plt.show()
```

```python
evaluate_alg(X_train, X_test, y_train, y_test, trees, coefs, eta)
```

```python
get_error_plot(n_trees, train_errors, test_errors)
```
